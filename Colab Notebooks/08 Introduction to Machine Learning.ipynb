# __Introduction to cost function Mean Squared Error (MSE)__
* Straight line equation used in __Machine learning__ is $\boldsymbol{y = wx + b}$ instead of $\boldsymbol{y = mx + c}$
* Erros in prediction is calcuatled by $\boldsymbol{(y_{\text{pred}} - y)}$
* To avoid negative value it is squared $\boldsymbol{(y_{\text{pred}} - y)^2}$
* Error calculation General form $\boldsymbol{(y_{\text{pred}}^{(i)} - y^{(i)})^2}$
* Error calculation for first point $\boldsymbol{(y_{\text{pred}}^{(0)} - y^{(0)})^2}$
* Error calculation for second point $\boldsymbol{(y_{\text{pred}}^{(1)} - y^{(1)})^2}$
* Error calculation for 10th point $\boldsymbol{(y_{\text{pred}}^{(9)} - y^{(9)})^2}$
* Error calculation for last point $\boldsymbol{(y_{\text{pred}}^{(1459)} - y^{(1459)})^2}$
* Summing all the errors $\boldsymbol{(y_{\text{pred}}^{(0)} - y^{(0)})^2} + {(y_{\text{pred}}^{(1)} - y^{(1)})^2} + {(y_{\text{pred}}^{(2)} - y^{(2)})^2 ....... {(y_{\text{pred}}^{(1459)} - y^{(1459)})^2}} $
* Taking the mean of the sum $\boldsymbol{\frac{1}{1460} \left[ (y_{\text{pred}}^{(0)} - y^{(0)})^2 + (y_{\text{pred}}^{(1)} - y^{(1)})^2 + (y_{\text{pred}}^{(2)} - y^{(2)})^2 + \dots + (y_{\text{pred}}^{(1459)} - y^{(1459)})^2 \right]}$
* Sum of errors of all the points $\boldsymbol{\frac{1}{1460} \Sigma_{i=0}^{1459} (y_{\text{pred}}^{(i)} - y^{(i)})^2}$
* Mean Sqaure Error  $\boldsymbol {MSE =} {\frac{1}{m} \Sigma_{i}^{m} (y_{\text{pred}}^{(i)} - y^{(i)})^2}$

# __Key Parameters or coefficient Slope w and y-intercept b__
* We can write our straight line equation as $\boldsymbol{f_{w,b}(x) = w \cdot x + b}$ instead of $\boldsymbol{y = w \cdot x + b}$
* or simply it can be written $\boldsymbol{f(x) = w \cdot x + b}$
* $\boldsymbol{w,b}$ are called __parameters or coefficients or weights__ in machine learning
* Similary MSE can be expressed as $\boldsymbol{J(w, b) =\frac{1}{m} \Sigma_{i}^{m} (y_{\text{pred}}^{(i)} - y^{(i)})^2}$
* As $\boldsymbol{y_{\text{pred}} = f_{w,b}(x^{(i)})}$ therefore cost function will change to $\boldsymbol{J(w, b) =\frac{1}{m} \Sigma_{i}^{m} (f_{w,b}(x^{(i)}) - y^{(i)})^2}$
* __Our goal is to find best value of slope w which will minimize our MSE__
* For simplification we have set y-intecept b = 0

```python
# Define x values
x = range(5)  # x values from 0 to 4

# Define the conditions
conditions = [
    {"w": 0, "b": 2, "label": "w=0, b=2"},
    {"w": 1, "b": 0, "label": "w=1, b=0"},
    {"w": 1, "b": 1, "label": "w=1, b=1"}
]

# Create subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns

# Generate each subplot
for i, condition in enumerate(conditions):
    w = condition["w"]
    b = condition["b"]
    label = condition["label"]
    y = [w * val + b for val in x]  # Straight-line equation

    axs[i].plot(x, y, label=label)  # Plot the line
    axs[i].set_title(label)  # Use the label as the title
    axs[i].set_xlim(0, 5)  # Set x-axis limit from 0 to infinity
    axs[i].set_ylim(0, 5)  # Set y-axis limit from 0 to infinity
    axs[i].grid(True)  # Enable grid
    axs[i].legend(loc="best", fontsize=10)  # Add the label as legend

# Set overall layout
plt.tight_layout()
plt.show()
```

```python
mean_squared_errors = {}  # Initialize a dictionary to store mean squared errors

# Single loop to handle all operations
for w in range(-2000, 2001, 50):  # w values from -2000 to 2000, step 50
    # Generate predicted values
    col_name_pred = f'y_pred_w{w}'  # Predicted value column
    model_dataframe[col_name_pred] = model_dataframe['x'] * w  # Calculate y_pred = w * x

    # Calculate squared errors
    col_name_error = f'squared_error_w{w}'  # Squared error column
    model_dataframe[col_name_error] = (model_dataframe[col_name_pred] - model_dataframe['y']) ** 2  # Squared error calculation

    # Calculate mean squared error
    mean_squared_errors[w] = model_dataframe[col_name_error].mean()  # Compute mean squared error

# Create a new DataFrame for mean squared errors
mean_squared_error_df = pd.DataFrame(
    list(mean_squared_errors.items()),
    columns=['Slope (w)', 'Mean Squared Error']
)

mean_squared_error_df.head(50)
```

```python
# Sort slopes and plot
mean_squared_error_df.sort_values(by='Slope (w)', inplace=True)  # Sort by slope
mean_squared_error_df.plot(x='Slope (w)', y='Mean Squared Error', kind='line')
plt.title('Mean Squared Error vs. Slope (w)')
plt.xlabel('w')
plt.ylabel(r'$\boldsymbol{J(w)}$')
plt.grid(True)
plt.show()
```

* The derivative for $\boldsymbol{w}$ is $\boldsymbol{\frac{\partial J(w, b)}{\partial w}}$
* The derivative for $\boldsymbol{b}$ is $\boldsymbol{\frac{\partial J(w, b)}{\partial b}}$
* $\boldsymbol{w = w - \frac{\partial J(w, b)}{\partial w}}$
* $\boldsymbol{b = b - \frac{\partial J(w, b)}{\partial b} }$
* $\boldsymbol{w_{\text{temp}} = w - \frac{\partial J(w, b)}{\partial w}}$
* $\boldsymbol{b_{\text{temp}} = b - \frac{\partial J(w, b)}{\partial b} }$
* $\boldsymbol{w = w_{\text{temp}}}$
* $\boldsymbol{b = b_{\text{temp}}}$

# __What is Gradient Descent__
* The __gradient__ is the __numerical value__ of the slope of a tangent line on a curve.  
* It measures how __steep__ the curve is at a specific point.  
* __Gradients descent__ guide optimization by showing the direction and rate of change in functions.  
* __Negative gradients__ occur on the left side of a curve, while __positive gradients__ occur on the right.  
* __Gradient descent__ uses gradients to __iteratively__ minimize the __cost function__ step by step.  
* At the __minimum value__ of the curve, the gradient becomes __zero__, indicating no slope.  

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for the curve
w = np.linspace(-10, 10, 100)
j_w = w**2  # Example of a quadratic curve

# Define tangent points and slopes
w_neg = -5  # Left side of the curve for negative slope
w_pos = 5   # Left side of the curve for positive slope
negative_slope = 2 * w_neg  # Slope at w = -5
positive_slope = 2 * w_pos  # Slope at w = 5

# Generate shorter tangent lines
tangent_length = 2
w_tangent_neg = np.linspace(w_neg - tangent_length, w_neg + tangent_length, 10)
w_tangent_pos = np.linspace(w_pos - tangent_length, w_pos + tangent_length, 10)

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# First subplot: negative slope
axs[0].plot(w, j_w, label='Curve of J(w)')
axs[0].plot(w_neg, w_neg**2, 'ro', label='Point of Tangency')
axs[0].plot(w_tangent_neg, negative_slope * (w_tangent_neg - w_neg) + w_neg**2,
            label='Tangent Line (-ve Slope)', linestyle='--')
axs[0].set_title('Negative Slope')
axs[0].set_xlabel('w')
axs[0].set_ylabel('J(w)')
axs[0].legend()
axs[0].text(w_neg - 1, w_neg**2 + 10, '-ve slope', fontsize=12, color='blue')

# Second subplot: positive slope
axs[1].plot(w, j_w, label='Curve of J(w)')
axs[1].plot(w_pos, w_pos**2, 'ro', label='Point of Tangency')
axs[1].plot(w_tangent_pos, positive_slope * (w_tangent_pos - w_pos) + w_pos**2,
            label='Tangent Line (+ve Slope)', linestyle='--')
axs[1].set_title('Positive Slope')
axs[1].set_xlabel('w')
axs[1].set_ylabel('J(w)')
axs[1].legend()
axs[1].text(w_pos - 1, w_pos**2 + 10, '+ve slope', fontsize=12, color='green')

# Display the plots
plt.tight_layout()
plt.show()
```

# __Learning rate__
- To overcome gradient descent oscillations around the minimum, the learning rate is introduced.  
- The learning rate helps scale down derivative impact, allowing controlled and stable parameter updates.  
- Proper application of a learning rate ensures convergence to the minimum without overshooting.  
- A small learning rate slows down progress, while a large one can cause instability.  
- Fine-tuning the learning rate accelerates convergence and ensures efficient gradient descent optimization.


* Learnign rate application to Gradient Descent algorithm
  * $w = w - \alpha \cdot \frac{\partial J(w, b)}{\partial w}$
  * $b = b - \alpha \cdot \frac{\partial j(w, b)}{\partial b}$

```python
def plot_mse_vs_iterations(mse_history):
    """
    Plots MSE vs Iterations.

    This function creates a line plot to visualize the progression of the
    Mean Squared Error (MSE) over the iterations of gradient descent.

    Parameters:
    mse_history : numpy.ndarray
        The Mean Squared Error values recorded across all iterations of gradient descent.

    Returns:
    None
        Displays the plot of MSE vs Iterations.
    """
    iterations = range(1, len(mse_history) + 1)  # Create iteration numbers (1-based indexing)

    plt.figure(figsize=(8, 6))
    plt.plot(iterations, mse_history, marker='o', linestyle='-', color='blue')
    plt.xlabel("Iterations", fontsize=12)
    plt.ylabel("Mean Squared Error (MSE)", fontsize=12)
    plt.title("MSE vs Iterations", fontsize=14)
    plt.grid()
    plt.show()
```

```python
def linear_regression(data, learning_rate=0.1, training_iter=100, w_initail=-50.0, verbose=False):
    """
    Performs linear regression using gradient descent and returns unscaled predictions (y_pred).

    This function applies gradient descent to find the optimal weights and bias
    for a given dataset. It includes feature scaling, dynamically tracks the
    history of weights, bias, and MSE during the training process, and calculates
    unscaled predictions using the best weights and bias.

    Parameters:
    data : pandas.DataFrame
        The input dataset containing features and the target column.
    learning_rate : float, optional (default=0.1)
        The step size used for updating weights and bias during gradient descent.
    training_iter : int, optional (default=100)
        The number of iterations to run the gradient descent process.

    Returns:
    tuple:
        - best_w : numpy.ndarray
            The best weights obtained at the iteration with the lowest MSE.
        - best_b : float
            The best bias obtained at the iteration with the lowest MSE.
        - best_mse : float
            The lowest Mean Squared Error achieved during training.
        - y_pred : numpy.ndarray
            The unscaled predictions based on the best weights and bias.
    """
    df_x = data.copy()

    # Reset DataFrame if 'y_pred' exists

    df_x = df_x.loc[:, ~df_x.columns.str.contains('_pred')]
    target = df_x.columns[-1]

    data_dim = df_x.shape

    # Feature scaling
    for col in df_x.columns:
        df_x[col + '_scaled'] = (df_x[col] - df_x[col].mean()) / df_x[col].std()

    w = np.full((data_dim[1] - 1), w_initail)

    if verbose:
        print(f'{w=}')

    b = 0

    w_history = np.zeros((training_iter, data_dim[1] - 1))  # Store weights (iterations, features)
    b_history = np.zeros(training_iter)  # Store bias values
    mse_history = np.zeros(training_iter)  # Store MSE values


    # Select columns with '_scaled' in their name except the last column
    features = df_x.columns[df_x.columns.str.contains('_scaled')][:-1]
    target_scaled = df_x.columns[df_x.columns.str.contains('_scaled')][-1]

    # Gradient descent
    for iteration in range(training_iter):

        df_x[target+'_pred_scaled'] = df_x[features].dot(w) + b  # Compute scaled predictions
        df_x['error'] = df_x[target+'_pred_scaled'] - df_x[target+'_scaled']  # Compute error

        w_gradients = 2 * df_x['error'].dot(df_x[features].values) / data_dim[0]
        b_gradients = 2 * df_x['error'].mean()

        # Update weights and bias
        w -= learning_rate * w_gradients
        b -= learning_rate * b_gradients
        mse = (df_x['error'] ** 2).mean()

        if verbose:
            print(f"{df_x[['error',target+'_pred_scaled']].head()=}")
            print(f'{w_gradients=}')
            print(f'{b_gradients=}')
            print(f'{w=}')
            print(f'{b=}')
            print(f'{mse=}')

        w_history[iteration] = w
        b_history[iteration] = b
        mse_history[iteration] = mse

    # Find best parameters
    best_index = np.argmin(mse_history)
    best_w = w_history[best_index]
    best_b = b_history[best_index]
    best_mse = mse_history[best_index]

    # Compute scaled predictions for the target variable
    df_x[target+'_pred_scaled'] = df_x[features].dot(best_w) + best_b

    # Compute unscaled predictions
    y_mean = data[target].mean()  # Mean of original target column
    y_std = data[target].std()    # Standard deviation of original target column
    y_pred = ((df_x[target+'_pred_scaled'] * y_std) + y_mean).astype('int')  # Unscale predictions as int

    if verbose:
        print(df_x.head())

    # Plot MSE vs Iterations
    plot_mse_vs_iterations(mse_history)

    return best_w, best_b, best_mse, y_pred
```

```python
def extract_feature_info(feature_name, file_lines):
    """
    Extracts the description and details of a specified feature from a list of file lines.

    Parameters:
        feature_name (str): The name of the feature to extract.
        file_lines (list): The lines of the file as a list of strings.

    Returns:
        tuple: A tuple containing the feature description (str) and details (str).
    """
    description = None
    details = []  # To collect lines with feature details
    is_feature_section = False

    for idx, line in enumerate(file_lines):
        line = line.rstrip()  # Remove trailing whitespace

        # Check if the line is the feature name
        if line.startswith(f"{feature_name}:"):
            # Extract the description for the feature
            description = line.split(":", 1)[1].strip()
            is_feature_section = True  # Start capturing details from subsequent lines
            continue

        # If capturing details, stop when encountering a new feature
        if is_feature_section:
            # A new feature starts if ":" is present AND the line does NOT start with a space
            if ":" in line and not line.startswith(" "):
                break

            # Otherwise, add non-empty lines to details
            if line.strip():  # Skip empty lines
                details.append(line.strip())

    # Combine details into a single string with proper formatting
    details_str = "\n".join(details)
    return feature_name, description, details_str
```

# __Exploring Feature Scaling Methods__
* Standard Scaling or Z-Score Normalization $\mathbf{X_{\text{scaled}} = \frac{X - \mu}{\sigma}}$  Where ${\mu}$ = mean and ${ \sigma}$ = standard  deviation
*  Mean normalizatoin $\mathbf{X_{\text{normalized}} = \frac{X - \mu}{X_{\text{max}} - X_{\text{min}}}}$ Where ${\mu}$ = mean
* Max normalization $\mathbf{X_{\text{normalized}} = \frac{X}{X_{\text{max}}}}$
* Normal mean normal distribution

```python
# Generate synthetic data with a perfect Gaussian distribution for smooth bell curve
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=200)  # Normal distribution
data = np.concatenate([data, np.random.uniform(1, 20, size=20)])  # Add small tail
data = np.concatenate([data, np.random.uniform(100, 200, size=20)])  # Add large head

# Apply scaling methods
# Z-Score Normalization
mean_value = np.mean(data)
std_deviation = np.std(data)
z_score_scaled = (data - mean_value) / std_deviation

# Min-Max Scaling
min_value = np.min(data)
max_value = np.max(data)
min_max_scaled = (data - min_value) / (max_value - min_value)

# Max Normalization
max_value = np.max(data)
max_normalized = data / max_value

# Plot the original and scaled distributions
plt.figure(figsize=(14, 10))

# Original Data Density Distribution
plt.subplot(2, 2, 1)
density, bins = np.histogram(data, bins=30, density=True)
plt.plot((bins[:-1] + bins[1:]) / 2, density, color='blue', label="Original Data")
plt.title("Original Data Distribution")
plt.xlabel("Feature Values")
plt.ylabel("Density")
plt.legend()

# Z-Score Normalized Data Density Distribution
plt.subplot(2, 2, 2)
density, bins = np.histogram(z_score_scaled, bins=30, density=True)
plt.plot((bins[:-1] + bins[1:]) / 2, density, color='green', label="Z-Score Normalized Data")
plt.title("Z-Score Normalized Distribution")
plt.xlabel("Feature Values")
plt.ylabel("Density")
plt.legend()

# Min-Max Scaled Data Density Distribution
plt.subplot(2, 2, 3)
density, bins = np.histogram(min_max_scaled, bins=30, density=True)
plt.plot((bins[:-1] + bins[1:]) / 2, density, color='orange', label="Min-Max Scaled Data")
plt.title("Min-Max Scaled Distribution")
plt.xlabel("Feature Values")
plt.ylabel("Density")
plt.legend()

# Max Normalized Data Density Distribution
plt.subplot(2, 2, 4)
density, bins = np.histogram(max_normalized, bins=30, density=True)
plt.plot((bins[:-1] + bins[1:]) / 2, density, color='red', label="Max Normalized Data")
plt.title("Max Normalized Distribution")
plt.xlabel("Feature Values")
plt.ylabel("Density")
plt.legend()

plt.tight_layout()
plt.show()
```

# __Polynomial regression and Feature Engineering__
* Linear Polynomial (Degree 1) $\mathbf{P(x) = a_1x + a_0}$
* Quadratic Polynomial (Degree 2) $\mathbf{P(x) = a_2x^2 + a_1x + a_0}$
* Cubic Polynomial (Degree 3) $\mathbf{P(x) = a_3x^3 + a_2x^2 + a_1x + a_0}$
* General Polynomial (Degree n) $\mathbf{P(x) = a_nx^n + a_{n-1}x^{n-1} + \dots + a_1x + a_0}$
- A polynomial is a mathematical expression combining variables raised to powers through basic operations.
- __Polynomial regression__ captures curved relationships, extending the concept to fit data points effectively.
- __Feature engineering__ transforms features, like squaring or cubing, to reveal hidden non-linear trends in data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Generate synthetic dataset with more complexity
np.random.seed(42)
X = np.random.uniform(-10, 10, 100).reshape(-1, 1)  # Independent variable
y = 0.1 * (X**3) - 0.5 * (X**2) + 2 * X + np.random.normal(0, 15, 100).reshape(-1, 1)  # Cubic relationship with noise

# Create a single figure for all plots
plt.figure(figsize=(16, 12))

# Plot 1: Original Dataset
plt.subplot(2, 2, 1)
plt.scatter(X, y, color='blue', label='Original Data')
plt.title("Original Dataset")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()

# Plot 2: Linear Fit
model_linear = LinearRegression()
model_linear.fit(X, y)
y_pred_linear = model_linear.predict(X)
plt.subplot(2, 2, 2)
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred_linear, color='red', label='Linear Fit')
plt.title("Linear Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()

# Plot 3: Quadratic Polynomial Fit (Degree 2)
model_quad = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model_quad.fit(X, y)
X_sorted = np.sort(X, axis=0)
y_pred_quad = model_quad.predict(X_sorted)
plt.subplot(2, 2, 3)
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X_sorted, y_pred_quad, color='green', label='Quadratic Fit')
plt.title("Polynomial Regression (Degree 2)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()

# Plot 4: Cubic Polynomial Fit (Degree 3)
model_cubic = make_pipeline(PolynomialFeatures(degree=3), LinearRegression())
model_cubic.fit(X, y)
y_pred_cubic = model_cubic.predict(X_sorted)
plt.subplot(2, 2, 4)
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X_sorted, y_pred_cubic, color='orange', label='Cubic Fit')
plt.title("Polynomial Regression (Degree 3)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()

# Show all plots
plt.tight_layout()
plt.show()
```

# __Equations in Vector form__
* Signle feature and single point straight line equation
  * $\mathbf{f_{w,b}(x) = w \cdot x + b}$
* Multi features and muti point straght line equation
  * $\mathbf{f_{\vec{w}, b}(\vec{x}^{(i)}) = \left( \sum_{j=1}^{n} w_j x_j^{(i)} \right) + b}$
    * ${i = 1, 2, 3, ..., m}$ to represent mulitple points
    * ${j = 1, 2, 3, ..., n}$ to represent mulitple features
* Vector form of the equation
  * $\mathbf{f_{\vec{w}, b}(\vec{x}) = \vec{w} \cdot \vec{x} + b}$
* Gradient descent for single feature
  * $\mathbf{w = w - \alpha \cdot \frac{2}{m} \sum_{i=1}^{m} \left( f_{w,b}(x^{(i)}) - y^{(i)} \right) \cdot x^{(i)}}$
    * Where $\frac{\partial J(w,b)}{\partial w} = \frac{1}{m}\sum_{i=1}^{m}\left( f_{w,b}(x^{(i)}) - y^{(i)} \right)x^{(i)}$
  * $\mathbf{b = b - \alpha \cdot \frac{2}{m} \sum_{i=1}^{m} \left( f_{w,b}(x^{(i)}) - y^{(i)} \right)}$
    * Where $\frac{\partial J(w,b)}{\partial b} = \frac{1}{m}\sum_{i=1}^{m}\left( f_{w,b}(x^{(i)}) - y^{(i)} \right)$
  * Simultaneous update of ${w}$ and ${b}$
* Gradient descent for n features where n > 1
  * For ${j = 1}$ $\mathbf{w_1 = w_1 - \alpha \cdot \frac{2}{m} \sum_{i=1}^{m} \left( f_{\vec{w}, b}(\vec{x}^{(i)}) - y^{(i)} \right) \cdot x_1^{(i)}}$
  * For ${j = n}$ $\mathbf{w_n = w_n - \alpha \cdot \frac{2}{m} \sum_{i=1}^{m} \Bigl( f_{\vec{w}, b}(\vec{x}^{(i)}) - y^{(i)} \Bigr) \cdot x_n^{(i)}}$
  * $\mathbf{b = b - \alpha \cdot \frac{2}{m}\sum_{i=1}^{m} \Bigl( f_{\vec{w}, b}(\vec{x}^{(i)}) - y^{(i)} \Bigr)}$
  * Simultaneous update of ${w_j = 1, 2, 3,..., n}$ and ${b}$
