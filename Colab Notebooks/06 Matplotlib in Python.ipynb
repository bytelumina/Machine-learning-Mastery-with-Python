```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
data = np.random.normal(size=100)

# Calculate KDE using numpy
x = np.linspace(min(data), max(data), 100)
density = np.zeros_like(x)
bandwidth = 1.06 * data.std() * np.power(len(data), -1/5)  # Silverman's rule of thumb

for xi in x:
    density += np.exp(-(data - xi)**2 / (2 * bandwidth**2))

density /= (bandwidth * np.sqrt(2 * np.pi) * len(data))

# Create the KDE plot using a line plot and fill_between
plt.plot(x, density, label='KDE')
plt.fill_between(x, density, alpha=0.5)

# Add a title and labels
plt.title('Kernel Density Estimate (KDE) Plot')
plt.xlabel('Value')
plt.ylabel('Density')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

# __Figure, subplot and subplots__
#### Matplotlib's __figure__ object is the top-level container for creating complex plots manually.
#### The __plt.subplot(nrows, ncols, index)__ function creates a single subplot within a figure by specifying its position.
#### The __plt.subplots(nrows, ncols)__ function creates multiple subplots and returns a figure and axes objects.
#### Customizing subplot positions and sizes is easier with subplots due to __axes__ object access.
#### __Subplot and subplots__ enable detailed visualizations, comparing multiple datasets within one figure.
#### Use of __for__ loop can speed up the work
#### Choosing between subplot and subplots depends on specific needs and customization preferences.

```python
import matplotlib.pyplot as plt

# Create a figure
fig = plt.figure()

# Add three subplots to the figure
plt.subplot(3, 1, 1)
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.title('Plot 1')

plt.subplot(3, 1, 2)
plt.plot([1, 2, 3, 4], [30, 25, 20, 10])
plt.title('Plot 2')

plt.subplot(3, 1, 3)
plt.plot([1, 2, 3, 4], [15, 18, 22, 28])
plt.title('Plot 3')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure with multiple subplots (2 rows, 2 columns)
fig, axs = plt.subplots(2, 2)

# Plot data on each subplot using axes objects
axs[0, 0].plot([1, 2, 3, 4], [10, 20, 25, 30])
axs[0, 0].set_title('Plot 1')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

axs[0, 1].plot([1, 2, 3, 4], [30, 25, 20, 10])
axs[0, 1].set_title('Plot 2')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')

axs[1, 0].plot([1, 2, 3, 4], [15, 18, 22, 28])
axs[1, 0].set_title('Plot 3')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')

axs[1, 1].plot([1, 2, 3, 4], [5, 15, 10, 20])
axs[1, 1].set_title('Plot 4')
axs[1, 1].set_xlabel('X-axis')
axs[1, 1].set_ylabel('Y-axis')

# Display the plot
plt.tight_layout()
plt.show()

```

# __Colors, Markers, and Line Styles in Matplotlib__
#### Matplotlib provides extensive options for customizing plot appearance, such as colors, markers, and line styles.
#### Colors help distinguish different lines and data points, enhancing plot clarity and aesthetics.
#### Markers represent individual data points with various styles, like circles, squares, and triangles.
#### Line styles define line appearance, offering options like solid, dashed, dash-dot, and dotted lines.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sinh(x)

# Create a figure with multiple subplots (2 rows, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot data on each subplot with different styles
axs[0, 0].plot(x, y1, color='b', marker='o', linestyle='-')
axs[0, 0].set_title('Sine Function')
axs[0, 0].set(xlabel='X-axis', ylabel='Y-axis')

axs[0, 1].plot(x, y2, color='g', marker='s', linestyle='--')
axs[0, 1].set_title('Cosine Function')
axs[0, 1].set(xlabel='X-axis', ylabel='Y-axis')

axs[1, 0].plot(x, y3, color='r', marker='^', linestyle='-.')
axs[1, 0].set_title('Tangent Function')
axs[1, 0].set(xlabel='X-axis', ylabel='Y-axis')

axs[1, 1].plot(x, y4, color='m', marker='d', linestyle=':')
axs[1, 1].set_title('Hyperbolic Sine Function')
axs[1, 1].set(xlabel='X-axis', ylabel='Y-axis')

# Display the plot

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sinh(x)

# Create a figure with multiple subplots (2 rows, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# List of y data for each subplot
y_data = [y1, y2, y3, y4]
titles = ['Sine Function', 'Cosine Function', 'Tangent Function', 'Hyperbolic Sine Function']
colors = ['b', 'g', 'r', 'm']
markers = ['o', 's', '^', 'd']
linestyles = ['-', '--', '-.', ':']

# Customizing all subplots using a loop
for ax, y, title, color, marker, linestyle in zip(axs.flat, y_data, titles, colors, markers, linestyles):
    ax.plot(x, y, color=color, marker=marker, linestyle=linestyle)
    ax.set_title(title)
    ax.set(xlabel='X-axis', ylabel='Y-axis')
    ax.set_xticks([0, 2, 4, 6, 8, 10])
    ax.set_xticklabels(['0', '2', '4', '6', '8', '10'])
    ax.set_yticks(np.linspace(y.min(), y.max(), 5))
    ax.set_yticklabels(np.round(np.linspace(y.min(), y.max(), 5), 2))

# Turning off axes for one of the subplots
axs[1, 0].axis('off')

# Display the plot
plt.tight_layout()
plt.show()
```

# __Annotation and drawings on Subplot__
#### The __annotate()__ function adds text annotations, specifying text, coordinates, and appearance.
#### Shapes like rectangles, circles, and polygons can be drawn using the __patches module__.
#### Combining annotations and drawings makes plots more informative and visually appealing.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()

# Plot data
ax.plot(x, y)

# Add annotations
ax.annotate('Max Point', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

ax.annotate('Min Point', xy=(3*np.pi/2, -1), xytext=(3*np.pi/2 + 1, -1.5),
             arrowprops=dict(facecolor='blue', shrink=0.05))

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()

# Plot data
ax.plot(x, y)

# Add rectangle
rect = patches.Rectangle((2, -1), 2, 2, linewidth=1, edgecolor='r', facecolor='none')
ax.add_patch(rect)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = np.exp(x)

# Create a figure with multiple subplots (2 rows, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot data on each subplot and add annotations and drawings
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('Sine Function')
axs[0, 0].annotate('Max Point', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 1.5),
                   arrowprops=dict(facecolor='black', shrink=0.05))
axs[0, 0].annotate('Min Point', xy=(3*np.pi/2, -1), xytext=(3*np.pi/2 + 1, -1.5),
                   arrowprops=dict(facecolor='blue', shrink=0.05))

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cosine Function')
axs[0, 1].annotate('Min Point', xy=(np.pi, -1), xytext=(np.pi - 1, -1.5),
                   arrowprops=dict(facecolor='black', shrink=0.05))

axs[1, 0].plot(x, y3)
axs[1, 0].set_title('Quadratic Function')
rect = patches.Rectangle((2, 4), 2, 10, linewidth=1, edgecolor='r', facecolor='none')
axs[1, 0].add_patch(rect)

axs[1, 1].plot(x, y4)
axs[1, 1].set_title('Exponential Function')
circle = patches.Circle((5, np.exp(5)), 1, linewidth=1, edgecolor='b', facecolor='none')
axs[1, 1].add_patch(circle)

# Display the plot
plt.tight_layout()
plt.show()
```

# __Saving plots to file__
#### Matplotlib can save plots to files, useful for sharing, reports, or future reference.
#### The __savefig()__ function allows specifying the filename and format for the saved plot.
#### Matplotlib supports various file formats, including __PNG, JPG, SVG, PDF, and more.__
#### Customize the saved plot with parameters like __dpi__ for resolution and __bbox_inches__ for bounding box.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a plot
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Save the plot as a PNG file
plt.savefig('sine_wave.png')

# Display the plot
plt.show()
```

# __Introduction to Seaborn__
#### Seaborn is built on Matplotlib and makes statistical visualizations simple and aesthetically appealing.
#### It integrates seamlessly with Pandas DataFrames, making data handling and plotting effortless.
#### The __tips dataset__ contains helpful variables like total bill, tip, gender, and smoking status.
#### __Histograms__ in Seaborn help us visualize distributions, including patterns in a numerical dataset.
#### __Boxplots__ effectively display data spread and outliers, assisting in comparative analysis of groups.
#### __Pairplots__ show pairwise relationships and distributions for variables, ideal for multivariate exploration.
#### __Heatmaps__ highlight correlations between variables with color intensity for quick pattern identification.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in "tips" dataset
data = sns.load_dataset("tips")

# Plotting a histogram for 'total_bill'
sns.histplot(data['total_bill'], kde=True, color='blue')
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot showing relationship between total_bill and tip
sns.scatterplot(x='total_bill', y='tip', hue='time', data=data, palette='viridis')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# Boxplot to compare tips by day
sns.boxplot(x='day', y='tip', data=data, palette='pastel')
plt.title('Tips by Day')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in "tips" dataset
data = sns.load_dataset("tips")

# Create a pairplot to visualize relationships between variables
sns.pairplot(data, diag_kind='kde', hue='sex', palette='coolwarm')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in "tips" dataset
data = sns.load_dataset("tips")

# Select only numeric columns for correlation
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_data.corr()

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
